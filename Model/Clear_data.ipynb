```python
import pandas as pd
```

```python
df = pd.read_csv("Data/wiki_data.csv", sep=';')
df
```

```python
to_delete = []
for index, row in df.iterrows():
    if len(row.text) < 150:
        to_delete.append(index)
```

```python
df = df.drop(to_delete)
```

```python
df.to_csv("Data/wiki_data.csv")
```
